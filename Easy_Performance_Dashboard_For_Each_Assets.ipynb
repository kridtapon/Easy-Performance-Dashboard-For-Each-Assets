```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

# Download historical Bitcoin data (adjust dates as needed)
btc_data = yf.download("BTC-USD", start="2014-01-01", end="2024-12-31")
btc_data.columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
btc_data = btc_data.resample("ME").agg({"Open":"first",
                                 "High":"max",
                                 "Low":"min",
                                 "Close":"last",
                                 "Volume":"sum"})

# Calculate monthly returns
btc_data['Monthly Return'] = btc_data['Close'].pct_change()

# Extract year and month from the DateTime index
btc_data['Year'] = btc_data.index.year
btc_data['Month'] = btc_data.index.month

# Group by Year and Month to get average monthly returns for each year
monthly_returns = btc_data.groupby(['Year', 'Month'])['Monthly Return'].mean().unstack()

# Plotting the heatmap with a red-to-green colormap
plt.figure(figsize=(12, 6))
sns.heatmap(monthly_returns, annot=True, cmap='RdYlGn', fmt='.2%', linewidths=0.5, cbar_kws={'label': 'Monthly Return'})
plt.title('Bitcoin Monthly Returns Heatmap by Year')
plt.ylabel('Year')
plt.xlabel('Month')
plt.xticks(ticks=np.arange(0, 12, 1), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Download S&P 500 historical data
symbol = "^SPX"
data = yf.download(symbol, start="2000-01-01", end="2024-12-31")
data.columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

data = data.resample("ME").agg({"Open":"first",
                                 "High":"max",
                                 "Low":"min",
                                 "Close":"last",
                                 "Volume":"sum"})

# Step 2: Calculate monthly returns
data['Month'] = data.index.month
data['Year'] = data.index.year

# Calculate monthly returns
data['Monthly Return'] = data['Close'].pct_change()

# Step 3: Pivot data for heatmap (Year vs Month)
pivot_data = data.pivot_table(index='Year', columns='Month', values='Monthly Return', aggfunc=np.mean)

# Step 4: Create heatmap with red to green color scale
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='RdYlGn', annot=True, fmt=".2%", linewidths=0.5, center=0)

# Adding labels and title
plt.title("S&P 500 Monthly Returns (2000-2023)", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Year", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
